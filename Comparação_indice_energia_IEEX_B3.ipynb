```python
# de acordo com resultado çor percebido um erro no arquivo csv. os "indices que quia participa" na verdade trata-se dos tikers das empresas
#já os indices participantes estavam em uma coluna em branco. em vista disso foi essa adequação

import pandas as pd

# Renomear as colunas
colunas = ['Empresa', 'Código', 'Ticker da Empresa', 'Indices Participantes']

# Abrir o arquivo CSV em modo leitura
with open(r'C:\Users\berna\Dropbox\B3_IEE\AcoesIndices_2023-11-08.csv', 'r') as arquivo:
    linhas = arquivo.readlines()

# Excluir as primeiras duas linhas (cabeçalho)
linhas = linhas[2:]

# Criar listas para armazenar os dados organizados
empresas = []
codigos = []
tickers = []
indices_participantes = []

# Processar cada linha
for linha in linhas:
    # Dividir a linha onde houver ponto e vírgula
    colunas_linha = linha.strip().split(';')
    
    # Verificar se a linha possui informações suficientes
    if len(colunas_linha) >= 4:
        # Adicionar as informações às listas correspondentes
        empresas.append(colunas_linha[0])
        codigos.append(colunas_linha[1])
        tickers.append(colunas_linha[2])
        indices_participantes.append(colunas_linha[3])

# Criar um DataFrame pandas com os dados organizados
df = pd.DataFrame({
    'Empresa': empresas,
    'Código': codigos,
    'Ticker da Empresa': tickers,
    'Indices Participantes': indices_participantes
})

# Ajustar a exibição para truncar os valores da coluna 'Indices Participantes'
pd.set_option('display.max_colwidth', 50)

# Exibir as primeiras linhas do DataFrame
print(df.head(10))

```

```python
# Adicionar ".SA" aos tickers na coluna 'Ticker da Empresa', se ainda não tiver
df['Ticker da Empresa'] = df['Ticker da Empresa'].apply(lambda x: x + '.SA' if not x.endswith('.SA') else x)

# Remover repetições de ".SA" nos tickers
df['Ticker da Empresa'] = df['Ticker da Empresa'].apply(lambda x: x.replace('.SA', '') + '.SA')

# Exibir as primeiras linhas do DataFrame após a modificação
df
```

```python
# Filtrar as linhas que contêm 'IEEX' em 'Indices Participantes'
df_ieex = df[df['Indices Participantes'].str.contains('IEEX')]

# Exibir o DataFrame filtrado
df_ieex
```

```python
import yfinance as yf
import pandas as pd


# Lista dos tickers das 16 empresas a partir do DataFrame df_ieex
tickers_df_ieex = df_ieex['Ticker da Empresa']

# Criar um DataFrame vazio para armazenar as cotações dos últimos 12 meses
df_cotacoes_12meses = pd.DataFrame()

# Obter o histórico de cotações dos últimos 12 meses para cada ticker
for ticker in tickers_df_ieex:
    try:
        data = yf.download(ticker, period="1y")
        df_cotacoes_12meses[ticker] = data['Adj Close']
    except Exception as e:
        print(f"Falha ao obter histórico de cotações para {ticker}: {str(e)}")

# Exibir o DataFrame com o histórico de cotações dos últimos 12 meses
print(df_cotacoes_12meses)

```

```python
import matplotlib.pyplot as plt

# Identificar as 8 empresas com cotações mais altas
empresas_mais_altas = df_cotacoes_12meses.mean().nlargest(8).index

# Identificar as 8 empresas com cotações mais baixas
empresas_mais_baixas = df_cotacoes_12meses.mean().nsmallest(8).index

# Configurar o tamanho da figura
plt.figure(figsize=(15, 10))

# Plotar as cotações para as 8 empresas mais altas
plt.subplot(2, 1, 1)
for empresa in empresas_mais_altas:
    plt.plot(df_cotacoes_12meses.index, df_cotacoes_12meses[empresa], label=empresa)

plt.title('Cotações das 8 Empresas com Maior Valor nos Últimos 12 Meses')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento Ajustado')

# Plotar as cotações para as 8 empresas mais baixas
plt.subplot(2, 1, 2)
for empresa in empresas_mais_baixas:
    plt.plot(df_cotacoes_12meses.index, df_cotacoes_12meses[empresa], label=empresa)

plt.title('Cotações das 8 Empresas com Menor Valor nos Últimos 12 Meses')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento Ajustado')

# Ajustar o layout para evitar sobreposição
plt.tight_layout()

# Exibir os gráficos
plt.show()
```
